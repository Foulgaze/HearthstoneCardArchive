```python
import json
import os
```

```python
problematic_sets = set([1646,1635,1637])
problematic_names = ["Core.json", "Legacy.json", "Classic.json"]
all_sets = []

for file_name in os.listdir("SetData"):
	if file_name not in problematic_names:
		all_sets.append(json.loads(open(f"SetData/{file_name}","r" ).read()))

with open("all_cards.json", "r") as all_cards:
	all_cards_json  = json.loads(all_cards.read())
	for card_name in all_cards_json:
		card_data = all_cards_json[card_name]
		if "cardSetId" not in card_data:
			card_data["cardSetId"] = 0
			continue
		if card_data["cardSetId"] in problematic_sets:
			found_card = False
			for card_set in all_sets:
				if card_data["id"] in card_set:
					card_data["cardSetId"] = card_set["id"]
					found_card = True
					break
			if not found_card:
				print(f"Could not find {card_data['name']}")
				card_data["cardSetId"] = 0
with open("all_cards.json", "w") as outfile:
	outfile.write(json.dumps(all_cards_json, indent=4))
```
