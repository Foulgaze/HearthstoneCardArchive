```python
import json
import random
```

```python
infile = open("output.txt", "r")
outfile = open("all_cards.json", "w")
data = json.loads(infile.read())
fields = [
    "id",
    "collectible",
    "classId",
    "multiClassIds",
    "spellSchoolId",
    "cardTypeId",
    "cardSetId",
    "rarityId",
    "manaCost",
	"name",
    "image",
    "flavorText",
    "keywordIds",
]
output_dict = {}
all_card_names = list(data.keys())
random.shuffle(all_card_names)
print(all_card_names)
collectible = set()
for key in all_card_names:
	card_data = data[key]
	# print(card_data)
	name = card_data["name"].upper()
	output_dict[name] = {}
	keys = card_data.keys()
	for field in fields:
		if field in keys:
			if field == "cardSetId" and card_data[field] == 3:
				card_data[field] = 1635
			else:
			    output_dict[name][field] = card_data[field]
outfile.write(json.dumps(output_dict, indent=4))
```
